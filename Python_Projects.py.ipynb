## Number Guessing Game:

```python
import random
import math
lower = int(input("Enter Lower bound:- "))
upper = int(input("Enter Upper bound:- "))
x = random.randint(lower, upper)
print("\n\tYou've only ",round(math.log(upper - lower + 1, 2))," chances to guess the integer!\n")
count = 0
while count < math.log(upper - lower + 1, 2):
    count += 1
    guess = int(input("Guess a number:- "))
    if x == guess:
        print("Congratulations you did it in ",count, " try")
        break
    elif x > guess:
            print("You guessed too small!")
    elif x < guess:
            print("You Guessed too high!")
if count >= math.log(upper - lower + 1, 2):
    print("\nThe number is %d" % x)
    print("\tBetter Luck Next time!")
```

## ToDo Application:

```python
tasks = []

def add_task(task):
    tasks.append(task)
    print(f"Task '{task}' added.")


def remove_task(task):
    if task in tasks:
        tasks.remove(task)
        print(f"Task '{task}' removed.")
    else:
        print(f"Task '{task}' not found.")


def list_tasks():
    if tasks:
        print("Tasks:")
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")
    else:
        print("No tasks found.")
        
        
def update_task(task, new_task):
    if task in tasks:
        index = tasks.index(task)
        tasks[index] = new_task
        print(f"Task '{task}' updated to '{new_task}'.")
       
    else:
        print(f"Task '{task}' not found.")
    
def remove_all(tasks):
    if len(tasks)==0:
        print("There is no task present from begining.")
    else:
        tasks.clear()
        print("all tasks have been removed.")
        print("="*50)
   


while True:
    print("\nOptions:")
    print("1. Add Task")
    print("2. Remove Task")
    print("3. List Tasks")
    print("4. Update Tasks")
    print("5.Remove all Tasks")
    print("6. Exit")

    choice = input("Enter your choice: ")
    
    print("="*50)

    if choice == '1':
        task = input("Enter the task: ")
        add_task(task)
    elif choice == '2':
        task = input("Enter the task to remove: ")
        remove_task(task)
    elif choice == '3':
        list_tasks()
        print("="*50)
    elif choice == '4':
        task=input("enter the task to update:")
        new_task=input("enter the new task description:")
        update_task(task,new_task)
    elif choice=='5':
        remove_all(tasks)
       
    elif choice == '6':
        print("Exiting the to-do list application.")
        break
    else:
        print("Invalid choice. Please try again.")
```

## Hotel Room Reservation Project using OOPS:

```python
class Hotel:
    def __init__(self, name, capacity):
        self.name = name
        self.capacity = capacity
        self.rooms = []

    def add_room(self, room):
        if len(self.rooms) < self.capacity:
            self.rooms.append(room)
            print(f"Room {room.number} added to {self.name}")
        else:
            print("No more rooms can be added. Maximum capacity reached.")

    def display_rooms(self):
        print(f"\nRooms in {self.name}:")
        for room in self.rooms:
            print(room)

class Room:
    def __init__(self, number, capacity, rate_per_night):
        self.number = number
        self.capacity = capacity
        self.rate_per_night = rate_per_night
        self.is_reserved = False

    def reserve(self):
        if not self.is_reserved:
            self.is_reserved = True
            print(f"Room {self.number} reserved successfully.")
        else:
            print(f"Room {self.number} is already reserved.")

    def cancel_reservation(self):
        if self.is_reserved:
            self.is_reserved = False
            print(f"Reservation for Room {self.number} canceled.")
        else:
            print(f"No reservation found for Room {self.number}.")

    def __str__(self):
        status = "Reserved" if self.is_reserved else "Available"
        return f"Room {self.number} - Capacity: {self.capacity} - Status: {status} - Rate/Night: ${self.rate_per_night}"

class Reservation:
    def __init__(self, guest_name, room, nights):
        self.guest_name = guest_name
        self.room = room
        self.nights = nights
        self.total_cost = room.rate_per_night * nights

    def __str__(self):
        return f"Reservation for {self.guest_name} - Room {self.room.number} - {self.nights} nights - Total Cost: ${self.total_cost}"

# Example Usage:

# Create Hotel
my_hotel = Hotel("Grand Hotel", 10)

# Add Rooms to Hotel
room101 = Room(101, 2, 150)
room102 = Room(102, 4, 250)

my_hotel.add_room(room101)
my_hotel.add_room(room102)

# Display Hotel Rooms
my_hotel.display_rooms()

# Make Reservations
reservation1 = Reservation("John Doe", room101, 3)
reservation1.room.reserve()

reservation2 = Reservation("Jane Smith", room102, 2)
reservation2.room.reserve()

# Display Reservations
print("\nReservations:")
print(reservation1)
print(reservation2)

# Cancel Reservation
reservation1.room.cancel_reservation()

# Display Updated Rooms
my_hotel.display_rooms()
```
